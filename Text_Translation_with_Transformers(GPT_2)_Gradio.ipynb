```python
#Creating interface with Gradio and launches
gradio.Interface(generate_text,  title='GPT-2',
                 #generate_text в†’ function to call when user inputs text, "textbox" в†’ supposed to be input component,output_texts в†’ output component
                 inputs=gradio.Textbox(label="Enter prompt"),   # Input box
                outputs=gradio.Textbox(label="Generated text"),
              description=" GPT-2 is a transformer-based language model by OpenAI that generates coherent, human-like text from a given prompt and is widely used for text generation and NLP tasks."


                 ).launch() #.launch() в†’ opens the app in your browser
```
