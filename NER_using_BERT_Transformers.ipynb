```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline #a high-level API from Hugging Face that makes running tasks (like sentiment analysis, NER, etc.) very easy.
tokenizer= AutoTokenizer.from_pretrained("dslim/bert-base-NER") #automatically loads the correct tokenizer for any pretrained model (in this case, for a BERT model)."dslim/bert-base-NER" is a fine-tuned BERT-base model for Named Entity Recognition trained by user dslim on Hugging Face.
model= AutoModelForTokenClassification.from_pretrained("dslim/bert-base-NER") #loads a BERT model that is specifically fine-tuned for token classification tasks (like NER, POS tagging, etc.).
```

```python
classify= pipeline("ner", model=model, tokenizer=tokenizer)
data= "Spiders (order Araneae) are air-breathing arthropods that have eight limbs, chelicerae with fangs generally able to inject venom,[2] and spinnerets that extrude silk.[3] They are the largest order of arachnids and rank seventh in total species diversity among all orders of organisms.[4][5] Spiders are found worldwide on every continent except Antarctica, and have become established in nearly every land habitat. As of June 2025, 53,034 spider species in 136 families have been recorded by taxonomists.[1] However, there has been debate among scientists about how families should be classified, with over 20 different classifications proposed since 1900"
```

```python
results=classify(data)
results
```
